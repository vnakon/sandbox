### Trying to calculate cost of different clothing combinations

```python
from itertools import product
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# May be you can duplicate some of the columns bellow in case if you can choose them twice
```

```python
LegsBaseLayer=['Thermal Underwear','Wool Longjohns']
LegsOuterwear=['Jeans','Combat Pants','Cargo Pants', 'Work Pants','Snow Pants','Deerskin Pants']
TorsoBaseLayer=['Hoodie', 'Plaid Shirt','Sweatshirt','Dress Shirt','Thin Wool Sweater','T-Shirt','Wool Shirt','Thick Wool Sweater','Fishermans Sweater','Cowichan Sweater']
TorsoOuterwear=['Sport Vest','Windbreaker','Down Vest','Light Shell','Ski Jacket','Simple Parka','Mackinaw Jacket','Urban Parka','Military Coat','Expedition Parka','Mariners Pea Coat','Old Fashioned Parka','Wolfskin Coat','Bearskin Coat']
FeetShoes=['Leather Shoes','Running Shoes','Trail Boots','Combat Boots','Deerskin Boots','Insulated Boots','Mukluks','Work Boots','Ski Boots']
FeetSocks=['Sports Socks','Wool Socks','Climbing Socks']
Hands=['Driving Gloves','Fleece Mittens','Wool Mittens','Work Gloves','Ski Gloves','Gauntlets']
Head=['Baseball Cap','Cotton Scarf','Cotton Toque','Wool Ear Wrap','Wool Scarf','Fleece Cowl','Balaclava','Wool Toque','Long Wool Scarf']
```

```python
# Making cartesian product

df=pd.DataFrame(list(product(LegsBaseLayer,LegsOuterwear,TorsoBaseLayer,TorsoOuterwear,FeetShoes,FeetSocks,Hands,Head)),columns=['LegsBaseLayer','LegsOuterwear','TorsoBaseLayer','TorsoOuterwear','FeetShoes','FeetSocks','Hands','Head'])
```

```python
# Prepearing dictionary with different parameters based on clothing name:
df_dictionary=pd.read_csv('clothingStats.csv')
```

```python
# Choosing 'WarmthBonus' for following calculations:
df_dictionary_warmbonus=df_dictionary[['ClothingItem','WarmthBonus']]
dict=df_dictionary_warmbonus.set_index('ClothingItem')['WarmthBonus'].to_dict()
```

```python
df_dictionary_warmbonus.head()
```

```python
# Getting list of categorical columns
data_types = df.dtypes 
cat_cols = list(data_types[data_types=='object'].index)
```

```python
# Applying values (WarmthBonus) for each columns:
df_all = pd.DataFrame()
for cat_col in cat_cols:
    df_all[cat_col] = df[cat_col]
    
for cat_col in cat_cols:
    df_all['WarmthBonus_'+cat_col]=df_all[cat_col].apply(lambda x: dict[x])
```

```python
#df_all.fillna(0)
df_all.head()
```

```python
# Summaryzing WarmthBonus by rows:
df_all['totalWarmthBonus']=df_all.sum(axis=1)
```

```python
df_all.head()
```

```python
df_all_sorted = df_all.sort('totalWarmthBonus', ascending=False)
```

```python
df_all_sorted.head()
```
